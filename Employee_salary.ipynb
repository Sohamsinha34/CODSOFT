```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
salary=pd.read_csv('Employee_Salary.csv')
salary
```

```python
sns.jointplot(x='Years of Experience',y='Salary',data=salary)
```

```python
sns.lmplot(x='Years of Experience',y='Salary',data=salary)
```

```python
X=salary[['Years of Experience']]
y=salary[['Salary']]
```

```python
X_train=X
y_train=y
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regressor=LinearRegression()
#after performing linear regression we are getting irrelevent result
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly_regressor=PolynomialFeatures(degree=4) #how many times the graph is cut off
```

```python
X_columns=poly_regressor.fit_transform(X_train)
```

```python
print(X_columns)
```

```python
regressor_poly=LinearRegression()
regressor_poly.fit(X_columns,y_train)
```

```python
print('Model Coefficients:m ',regressor_poly.coef_)
print('Model Coefficients:b ',regressor_poly.intercept_)
```
